```python
"""
Base-2 algorithm from class
need to print a vertical string of the remainders
user reads the remainders from bottom to top for the binary 
"""

"""
This function prints 0 or 1, depending on if a number is divisible by 2
Parameter: x, an integer
Returns: remainder, a string 
"""
def print_remainder(x):
    remainder = "0" 
    if( x % 2 == 1 ):
        remainder = "1"
    return remainder
```

```python
x = 2025 #for this example, we are converting 2025 to binary
            #in future, would revise this script to take in user input

#loop keeps dividing by 2 until x becomes 0
while (x > 0): 
    print(print_remainder(x))
    x = x // 2 #need floor division here because of odd numbers 
                 #ex. 253/2 should round to 126
```
