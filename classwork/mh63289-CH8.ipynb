```python
"""
Challenge 8:
Implement a function that will take a matrix as a parameter and returns True or False 
depending if the matrix is Diagonally Dominant or not.


Your output will look like this:
<display the matrix>
Is the matrix diagonally dominant? True
"""
import numpy as np 

def diagonal_dom(A) -> bool:
    """
    Checks if a matrix is diagonally dominant 
    (for each row, the absolute value of the diagonal element is >= 
    the sum of the absolute values of the other elements in that row)
    
    Args:
        A: matrix
        
    Returns: 
        True or False depending if the matrix is Diagonally Dominant 
    """
    n = A.shape[0]
    for i in range(n):
        diag = abs(A[i][i])
        row_sum = np.sum(np.abs(A[i])) - diag  # sum of the other elements
        
        if diag < row_sum:
            return False  # exits if one row fails the test
            
    return True

A = np.array([
	[0.582745, 0.48	, 0.10	, 0.  	, 0.  	], 
 [0.48	, 1.044129, 0.46	, 0.10	, 0.  	],
 [0.10	, 0.46	, 1.10431 , 0.44	, 0.10	],
 [0.  	, 0.10	, 0.44	, 0.963889, 0.42	],
 [0.  	, 0.  	, 0.10	, 0.42	, 0.522565]], dtype=float)

print(A)
print("Is the matrix diagonally dominant?", diagonal_dom(A))

```
