```python
import math 

def calc_hyp(a, b):
    h = math.sqrt(a*a + b*b)
    return h
```

```python
# Monte Carlo Pi (1x1 square, how many spots are in the unit circle?)
# Area of circle / area of quare = pi/4
# thus, pi = 4*( area of circle / area of quare )
# these areas are where we use Monte Carlo randomness 
# AI prompt in Copilot "how to generate a random number from -1 to 1 in python"

import random

def approximatePi(numSamples):
    numHits = 0
    for i in range(0, numSamples):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        
        if( x*x + y*y < 1):
            numHits = numHits + 1
            
    return 4*numHits/numSamples

approximatePi(50)
approximatePi(300)
```
