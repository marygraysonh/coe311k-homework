```python
"""
Challenge 9:
Part 1. 
"""

import numpy as np 


# Data points
X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])

# Stack the design matrix with a column of ones (for the intercept)
A = np.vstack([X, np.ones(len(X))]).T

# Solve using least squares
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]

print(f"Slope: {m}, Intercept: {c}")
```

# Plot each data point
# Plot the line y= .95x + .59

```python
import matplotlib.pyplot as plt

# generate lines to plot
x_line = np.linspace(min(X), max(X), 100)
y_line = m * x_line + c 

# plot
plt.scatter(X, Y, label="Data Points")
plt.plot(x_line, y_line, label=f" Best Fit of y = {m}x + {c}", color="blue")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Least Squares Fit (y = .95x + .59)")
plt.legend()
plt.grid(True)
plt.show()
```
