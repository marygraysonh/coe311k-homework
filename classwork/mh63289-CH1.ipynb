```python
i = 0
j = 0 
t = 0
z = np.empty([2001,2001])
for x in np.arange(-10,10,.01):
    i = i + 1
    j = 0
    for y in np.arange(-10,10,.01):
        j = j + 1
        z[i][j] = x*x + y*y
```

```python
def f1(x,y):
    return (x*x + y*y)

x = np.linspace(-10, 10, 2000)
y = np.linspace(-10, 10, 2000)

X,Y = np.meshgrid(x,y)
Z = f1(X,Y)
```

```python
fig1 = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='hot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(45, 0)
```

```python
x = np.linspace(0, 2*np.pi, 100) #setting x-range

plt.subplot(2,2,1) #row, column, active area 
plt.plot(x, np.sin(x), 'r')

plt.subplot(2,2,2)
plt.plot(x, 2*x, 'g')

plt.subplot(2,2,3)
plt.plot(x, np.cos(x), 'b')

plt.subplot(2,2,4)
plt.plot(x, x*x, 'y')

plt.show()
```
