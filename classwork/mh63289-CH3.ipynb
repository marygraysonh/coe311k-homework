```python
"""
Challenge 3: 
Run Monte Carlo Pi approximation 100 times with 1000 samples
for each run, calculate absolute error abs(x_true - x_approx)
and relative error abs((x_true - x_approx)/ x_true)
"""
import random
import numpy as np
import matplotlib.pyplot as plt


"""
This function approximates Pi with the Monte Carlo method  
Paramater: numSamples, the amount of times you want the function to 'guess'
Returns: a float, the approximation of Pi 
"""
def approximatePi(numSamples): 
    numHits = 0
    for i in range(0, numSamples):
        x = random.uniform(-1, 1) #random x coordinate
        y = random.uniform(-1, 1) #random y coordinate
        
        if( x*x + y*y < 1):    #checks to see if coordinates inside unit circle
            numHits = numHits + 1
            
    return 4*numHits/numSamples 
```

```python
#main code here 

x = 100

absolute_errors = [] #array to store absolute errors
relative_errors = [] #array to store relative errors

for i in range(0,100): #100 times
    x_approx = approximatePi(1000) #1000 samples each time 
    x_true = np.pi
    absolute_error = abs(x_true - x_approx) 
    relative_error = abs((x_true - x_approx))/ x_true
    absolute_errors.append(absolute_error) #store error values 
    relative_errors.append(relative_error)
```

```python
#plot both errors over 100 times here 

x = np.linspace(0,100,100)

plt.subplot(2,2,1) #row, column, active area 
plt.plot(x, absolute_errors, 'r')
plt.title('Absolute Error')

plt.subplot(2,2,2)
plt.plot(x, relative_errors, 'b')
plt.title('Relative Error')

plt.show()
```
