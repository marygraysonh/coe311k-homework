```python
# Approximate sin(x) using Taylor Series
import math
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def sin(x, terms) -> float:
    """
    Approximates sin(x) with Taylor Series
    Has capability of varying the amount of terms used for approximation

    Parameters: 
        x: specific value we want to approximate sin(x) at
        terms: amount of terms used in the Taylor approximation

    Returns:
        result: the approximate value of sin(x)
    """
    for n in range (0, terms):
        result = 0.
        term = ( ((-1)**n) * (x**(2*n + 1)) ) / (math.factorial(2*n + 1))
        result += term 
    return result 
```

```python
# Evaluate x for values 0-10, with steps of .10
x = np.linspace(0, 10, 101)
absolute_error = {}

# Test accuracy of 1-4 terms, compare approximations to exact value
for terms in range(1,5):
    taylor_sin = sin( x, terms )
    exact_sin = np.sin( x )
    absolute_error[terms] = np.abs(exact_sin - taylor_sin) 
    
```

```python
# Plot the truncation error as a function of the number of terms 

# 1 term plot
plt.figure()
plt.plot(x, absolute_error[1], 'r')
plt.title("Absolute Error of 1 Term")
plt.xlabel("x")
plt.ylabel("Absolute Error")
plt.show()

# 2 term plot
plt.figure()
plt.plot(x, absolute_error[2], 'b')
plt.title("Absolute Error of 2 Terms")
plt.xlabel("x")
plt.ylabel("Absolute Error")
plt.show()

# 3 term plot
plt.figure()
plt.plot(x, absolute_error[3], 'g')
plt.title("Absolute Error of 3 Terms")
plt.xlabel("x")
plt.ylabel("Absolute Error")
plt.show()

# 4 term plot
plt.figure()
plt.plot(x, absolute_error[4], 'y')
plt.title("Absolute Error of 4 Terms")
plt.xlabel("x")
plt.ylabel("Absolute Error")
plt.show()
```
