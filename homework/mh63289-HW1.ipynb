```python
"""
iterate the number of samples for your Monte Carlo Pi approx from 10 to 
1000, in steps of 20, and plot the error (
"""
import random
import numpy as np
import matplotlib.pyplot as plt


"""
This function approximates Pi with the Monte Carlo method  
Paramater: numSamples, the amount of times you want the function to 'guess'
Returns: a float, the approximation of Pi 
"""
def approximatePi(numSamples): 
    numHits = 0
    for i in range(0, numSamples):
        x = random.uniform(-1, 1) #random x coordinate
        y = random.uniform(-1, 1) #random y coordinate
        
        if( x*x + y*y < 1):    #checks to see if coordinates inside unit circle
            numHits = numHits + 1
            
    return 4*numHits/numSamples 

"""
Main code: iterates numSamples from 10 to 10000 in steps of 20,
plots the error (absolute value of calculated value of Pi - 
the actual value of Pi, to 15 digits past decimal pt)
"""
x = np.linspace(10, 10000, 20, dtype=int) #numSamples range,
#decided to add 'dtype=int' here to avoid errors with 'range()' above

errors = [] #array to store future values

for samp in x: #iterate through the samples
    guess = approximatePi(samp) 
    error = abs(guess - np.pi) 
    errors.append(error) #store error values 

plt.plot(x, errors, 'r')
plt.title('Error vs. Number of Samples')
plt.show()

#interesting observation: watching the graph fluctuate everytime I run it!
```
